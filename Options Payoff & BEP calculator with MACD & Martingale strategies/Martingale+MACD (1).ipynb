```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker_symbol='IRFC.NS'
```

```python
df=yf.download(ticker_symbol,start='2022-08-12',end='2024-07-12',interval='1h')
df
```

```python
def MACD(df):
    df['EMA_12']=df.Close.ewm(span=12).mean()
    df['EMA_26']=df.Close.ewm(span=26).mean()
    df['MACD']=df.EMA_12-df.EMA_26
    df['signal_line']=df.MACD.ewm(span=9).mean()
```

```python
MACD(df)
```

```python
df
```

```python
plt.figure(figsize=(14,3),dpi=360)
plt.plot(df.MACD,label='MACD',color='red')
plt.plot(df.signal_line,label='Signal',color='green')
plt.legend()
plt.show()
```

```python
long,short= [],[]
#first two row not taken as 0 value gives wrong indication
#both conditions to be checked to get whether crossover taken place or not
for i in range(2,len(df)):
    if df.MACD.iloc[i]>df.signal_line.iloc[i] and df.MACD.iloc[i-1]<df.signal_line[i-1]:
        long.append(i)
    elif df.MACD.iloc[i]<df.signal_line.iloc[i] and df.MACD.iloc[i-1]>df.signal_line[i-1]:
        short.append(i)
```

```python
df=df.reset_index()
```

```python
# Initialize the signal column with zeros
df['signal'] = 0

# Lists to store indices of long and short signals
long, short = [], []

# Iterate through the DataFrame, starting from the third row
for i in range(2, len(df)):
    if df.MACD.iloc[i] > df.signal_line.iloc[i] and df.MACD.iloc[i-1] < df.signal_line.iloc[i-1]:
        long.append(i)
        df.at[i, 'signal'] = 1  # Assign value 1 for long signals
    elif df.MACD.iloc[i] < df.signal_line.iloc[i] and df.MACD.iloc[i-1] > df.signal_line.iloc[i-1]:
        short.append(i)
        df.at[i, 'signal'] = 2  # Assign value 2 for short signals
    else:
        df.at[i, 'signal'] = 0  # Assign value 0 for no signals

# Verify the results
print(df[['MACD', 'signal_line', 'signal']])
```

```python
def SIGNAL():
    return df.signal

from backtesting import Strategy
class MyStrat(Strategy):
    initsize = 5
    mysize = initsize
    def init(self):
        super().init()
        self.signal1 = self.I(SIGNAL)

    def next(self):
        super().next()
        
        if(self.signal1>0 and len(self.trades)==0 and len(self.closed_trades)>0 and self.closed_trades[-1].pl < 0):
            #first condition ensures no previous trade is open and second condition ensures the P&L of last trade was negative meaning a loss hence we double the capital
            self.mysize=self.mysize*2
            #print(self.data.index, self.mysize)
            #print(self.closed_trades[-1], self.closed_trades[-1].pl)
            #print(self.equity)
            #print("-"*20)
        elif len(self.closed_trades)>0 and self.closed_trades[-1].pl > 0:
            #first condition ensures atleast one closed trade and secong one ensures last trade was a profit so we do not double the bet and keep it equal to initial
            self.mysize=self.initsize

        if self.signal1==2 and len(self.trades)==0:
            #defining stop loss and take profit ranges for sell signals
            sl1 = self.data.Close[-1] - 0.01
            tp1 = self.data.Close[-1] + 0.05
            self.buy(sl=sl1, tp=tp1,size=self.mysize)
        
        elif self.signal1==1 and len(self.trades)==0:
            #defining stop loss and take profit ranges for buy signals
            sl1 = self.data.Close[-1] + 0.01
            tp1 = self.data.Close[-1] - 0.05
            self.sell(sl=sl1, tp=tp1, size=self.mysize)
```

```python
from backtesting import Backtest

bt = Backtest(df, MyStrat, cash=100000, commission=.00)
stat = bt.run()
stat
```
