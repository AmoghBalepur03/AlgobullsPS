```python
def SIGNAL():
    return df.signal

from backtesting import Strategy
class MyStrat(Strategy):
    initsize = 1
    mysize = initsize
    def init(self):
        super().init()
        self.signal1 = self.I(SIGNAL)

    def next(self):
        super().next()
        
        if(self.signal1>0 and len(self.trades)==0 and len(self.closed_trades)>0 and self.closed_trades[-1].pl < 0):
            #first condition ensures no previous trade is open and second condition ensures the P&L of last trade was negative meaning a loss hence we double the capital
            self.mysize=self.mysize*2
            #print(self.data.index, self.mysize)
            #print(self.closed_trades[-1], self.closed_trades[-1].pl)
            #print(self.equity)
            #print("-"*20)
        elif len(self.closed_trades)>0 and self.closed_trades[-1].pl > 0:
            #first condition ensures atleast one closed trade and secong one ensures last trade was a profit so we do not double the bet and keep it equal to initial
            self.mysize=self.initsize

        if self.signal1==2 and len(self.trades)==0:
            #defining stop loss and take profit ranges for sell signals
            sl1 = self.data.Close[-1] - 0.05
            tp1 = self.data.Close[-1] + 0.05
            self.buy(sl=sl1, tp=tp1,size=self.mysize)
        
        elif self.signal1==1 and len(self.trades)==0:
            #defining stop loss and take profit ranges for buy signals
            sl1 = self.data.Close[-1] + 0.05
            tp1 = self.data.Close[-1] - 0.05
            self.sell(sl=sl1, tp=tp1, size=self.mysize)
```

```python
from backtesting import Backtest

bt = Backtest(df, MyStrat, cash=100000, commission=.00)
stat = bt.run()
stat
```
