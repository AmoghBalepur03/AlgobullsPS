The Code Snippet below is for getting the optimal combination of Interval-Period

```python
# Define the stock ticker symbol
ticker_symbol = 'RELIANCE.NS'  # .NS is for NSE (National Stock Exchange of India)

# Define intervals and periods
intervals = ['15m', '30m', '60m', '1h', '1d', '5d', '1wk', '1mo', '3mo']
periods = ['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']

# Initialize a dictionary to store the data
data_dict = {}

# MACD function
def MACD(df):
    df['EMA_12'] = df['Close'].ewm(span=12).mean()
    df['EMA_26'] = df['Close'].ewm(span=26).mean()
    df['MACD'] = df['EMA_12'] - df['EMA_26']
    df['signal_line'] = df['MACD'].ewm(span=9).mean()
    return df

# Function to calculate profit based on MACD strategy
def calculate_macd_profit(df):
    df = MACD(df)
    long, short = [], []

    for i in range(2, len(df)):
        if df['MACD'].iloc[i] > df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] < df['signal_line'].iloc[i-1]:
            long.append(i)
        elif df['MACD'].iloc[i] < df['signal_line'].iloc[i] and df['MACD'].iloc[i-1] > df['signal_line'].iloc[i-1]:
            short.append(i)

    # Adjusting for the time lag between signal generation and actual action
    Reallong = [i+1 for i in long if i+1 < len(df)]
    Realshort = [i+1 for i in short if i+1 < len(df)]

    # Ensure Reallong and Realshort are within bounds
    Reallong = [i for i in Reallong if i < len(df)]
    Realshort = [i for i in Realshort if i < len(df)]

    if not Reallong or not Realshort:
        return None

    # Since time lag is already taken into account, now we can go with the opening prices
    Buyprices = df['Open'].iloc[Reallong]
    Sellprices = df['Open'].iloc[Realshort]

    # Ensure indices are within bounds
    if len(Buyprices) == 0 or len(Sellprices) == 0:
        return None

    # Calculate profits and relative profits
    profits = [Sellprices.iloc[i] - Buyprices.iloc[i] for i in range(min(len(Buyprices), len(Sellprices)))]
    relprofits = [profits[i] / Buyprices.iloc[i] for i in range(len(profits))]

    if relprofits:
        return sum(relprofits) / len(relprofits)
    else:
        return None

# Loop through all combinations of intervals and periods to download data
for interval in intervals:
    for period in periods:
        # Skip combinations of (1m, 2m, 5m, 15m, 30m, 1h) with (1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max)
        if interval in ['1m', '2m', '5m', '15m', '30m', '1h'] and period in ['1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']:
            continue

        try:
            # Download the data
            data = yf.download(ticker_symbol, interval=interval, period=period)
            if not data.empty:
                data_dict[(interval, period)] = data
        except Exception as e:
            print(f"Error with interval {interval} and period {period}: {e}")

# Initialize a list to store results
results = []

# Calculate profit for each combination of interval and period
for (interval, period), data in data_dict.items():
    profit = calculate_macd_profit(data)
    if profit is not None:
        results.append({
            'Interval': interval,
            'Period': period,
            'Profit': profit
        })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Find the combination with the maximum profit
if not results_df.empty:
    max_profit_row = results_df.loc[results_df['Profit'].idxmax()]
    print("Best combination:")
    print(max_profit_row)
else:
    print("No valid data found.")
```
